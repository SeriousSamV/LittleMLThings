## Training and running a linear model using Scikit-Learn

```python
from sklearn.linear_model import LinearRegression

lifesat = pd.read_csv('https://github.com/ageron/data/raw/main/lifesat/lifesat.csv')
lifesat.head()
```

```python
X = lifesat[['GDP per capita (USD)']].values
y = lifesat['Life satisfaction'].values
lifesat.plot(kind='scatter', x='GDP per capita (USD)', y='Life satisfaction')
plt.show()
```

```python
model = LinearRegression()
model.fit(X, y)
```

inference/prediction using the trained/fitted model

```python
X_new = [[37_655.2]]
model.predict(X_new)
```
